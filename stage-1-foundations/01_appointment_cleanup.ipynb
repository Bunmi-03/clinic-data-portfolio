# P1 — Cedar Dental: Appointment Register Cleanup
**Scenario:** Cedar Dental has exported 515 appointment records. The data has missing values,
duplicate rows, and inconsistent appointment type labels. This notebook cleans the dataset
and produces a client-ready output file.

**Input:** `appointments_raw.csv`  
**Output:** `appointments_cleaned_YYYYMMDD.csv`  
**Libraries:** pandas only

```python
import pandas as pd
from datetime import datetime #  imports the datetime class from Python's built-in datetime module

df = pd.read_csv(r"C:\Users\clemo\Downloads\clinic-data-portfolio\data\appointments_raw.csv")
df
```

## 1. Data Inspection

```python
# Finding the shape of the dataframe
print("Shape:", df.shape)
```

```python
# Finding the datatypes of each set of data in the columns
print("Column types:")
print(df.dtypes)
```

```python
# Finds the null values in each data set
print("Null counts:")
print(df.isnull().sum())
```

```python
# Finds the sum of the duplicated values in each row
print("\nDuplicate rows:", df.duplicated().sum())
```

## 2. Fix: Drop duplicate Rows

```python
## Finds all the duplicate values
df[df.duplicated()]
```

```python
## Drops/Removes all the duplicate values
df = df.drop_duplicates().copy()
```

```python
df
```

```python
## Now we have removed all duplicate values from our dataframe
print("\nDuplicate rows:", df.duplicated().sum())
```

## 3. Fix appointment_type - Standardise Labels

```python
# Step 1: normalise casing and whitespace first
df['appointment_type'] = df['appointment_type'].str.lower().str.strip()
df
```

```python
# Step 2: check what you're working with after normalisation
print(df['appointment_type'].value_counts())
```

```python
 # Step 3: map dirty variants to 5 clean labels
correct_map = {
    'filling':  'filling',
    'fillig':        'filling',    # typo
    'check-up':      'check-up',
    'checkup':       'check-up',
    'check up':      'check-up',
    'chekup':        'check-up',   # typo
    'cleaning':      'cleaning',
    'clean':         'cleaning',
    'extraction':    'extraction',
    'consultation':  'consultation',
    'consult':       'consultation'
}

df['appointment_type'] = df['appointment_type'].map(correct_map)
df
```

```python
# Step 4: verify — should be 0 nulls and exactly 5 unique values
print(df['appointment_type'].value_counts()) 
print("Unmapped (nulls introduced):", df['appointment_type'].isnull().sum())
```

## 4. Fix: appointment_date — Parse to datetime

```python
# Standardise the dates and if you cannot parse a date then turn it into NaT
df['appointment_date'] = pd.to_datetime(df['appointment_date'], dayfirst=True, errors='coerce')
# Check how many bad dates got nulled out
print(df['appointment_date'].isnull().sum())
```

```python
# pd.to_datetime() converts dates into a full datetime object — date and time. Since your data has no time information, it just fills the time part with 00:00:00 as a default.
# If we want clean dates only then:
df['appointment_date'] = df['appointment_date'].dt.date
df
```

```python
# attended column has: yes, YES, Yes, 1, no, No, 0
print(df['attended'].value_counts())
```

```python
bool_map = {
    '0':  False,
    'yes' : True,
    'no': False,
    '1': True,
    'Yes': True,
    'YES': True,
    'No': False}

df['attended'] = df['attended'].map(bool_map)
df
```

## 5. Fix: patient_id — Flag Missing Records

```python
# Creating a new column called record_status. 'Incomplete' if pateient_id is null, therapist is null and attended is null. 'Complete' everywhere else
df['record_status'] = df.apply(
    lambda row: 'Incomplete' 
    if pd.isnull(row['patient_id']) or pd.isnull(row['therapist']) or pd.isnull(row['attended'])
    else 'Complete', axis=1
)


# To verify this
print(df['record_status'].value_counts())
```

## Summary

The raw appointment register (515 rows, 6 columns) had several data quality issues
that were identified, assessed, and resolved:

- **15 duplicate rows** removed — dataset reduced to 500 rows
- **17 appointment type variants** standardised to 5 clean labels:
  check-up, filling, cleaning, extraction, consultation
- **1 unparseable date** (`2024-00-10`) coerced to null — flagged for clinic review
- **attended column** standardised from mixed text/integer format to True/False boolean
- **76 missing attended values** retained as null — not filled to avoid false data
- **record_status column added** — 176 rows flagged Incomplete where patient_id,
  therapist, or attended data is missing
- No data was invented or filled — all decisions preserve data integrity
